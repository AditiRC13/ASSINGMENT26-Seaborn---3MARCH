Que 1: Name any five plots that we can plot using the Seaborn library. Also, state the uses of each plot.

```python
import seaborn as sns 
df=sns.load_dataset('iris')
sns.load_dataset('iris')
```

#plot 1
a. Distplot

(1)  Dist plot gives us the histogram of the selected continuous variable.

(2)  It is an example of a univariate analysis,
     We can change the number of bins i.e. number of vertical bars in a histogram

```python
sns.displot(df['sepal_length'])
sns.set(rc={'figure.figsize':(2,8)})
```

#plot2  b.Joint Plot

It is the combination of the distplot of two variables.

It is an example of bivariate analysis.We additionally obtain a scatter plot between the variable to reflecting their linear relationship. We can customize the scatter plot into a hexagonal plot, where, more the color intensity, the more will be the number of observations.

```python
import seaborn as sns

# For Plot 1
sns.jointplot(x = df['sepal_length'], y = df['petal_length'], kind = 'scatter',height=3,ratio=5,space=0.2)

# For Plot 2
sns.jointplot(x = df['sepal_length'], y = df['petal_length'],kind = 'hex',height=3,ratio=5,space=0.2)
```

#plot 3 c.Bar Plot

```python
import seaborn as sns
import matplotlib.pyplot as plt
sns.barplot(x = df['species'], y = df['petal_width'])
plt.rcParams['figure.figsize']=5,5
```

#plot 4 d.Strip Plot

Itâ€™s a plot between a continuous variable and a categorical variable.
It plots as a scatter plot but supplementarily uses categorical encodings of the categorical variable.

```python
import seaborn as sns
sns.stripplot(y = df['sepal_length'], x = df['species'],size=5,edgecolor='maroon')
```

#plot 5 e. Heat Map

In the given raw dataset â€˜dfâ€™, we have seven numeric variables. So, let us generate a correlation matrix between these seven variables.

```python
df.corr()
sns.heatmap(df.corr(), annot = True, cmap = 'viridis')
```

Que 2: Load the "fmri" dataset using the load_dataset function of seaborn. Plot a line plot using x =
"timepoint" and y = "signal" for different events and regions.

```python
sns.load_dataset('fmri')
```

```python
import numpy as np
import seaborn as sns; sns.set()
import matplotlib.pyplot as plt
fmri = sns.load_dataset("fmri")
fmri.sort_values('timepoint',inplace=True)
ax = sns.lineplot(x="timepoint", y="signal", data=fmri)
arr = np.ones(len(fmri))
arr[:300] = 0
arr[600:] = 2
fmri['background'] = arr
ax1 = sns.lineplot(x="timepoint", y="signal", hue="event", data=fmri)
ax2 = sns.lineplot(x="timepoint", y="signal", hue="region", data=fmri,color='maroon')
```
